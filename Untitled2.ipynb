```python
# Извлечение одного столбца
z = data[' Longitude']

# Преобразование строк в числа, игнорируя строки, которые не могут быть конвертированы
z = pd.to_numeric(z, errors='coerce')

# Удаление NaN значений
z = z.dropna().values

# Проверка на наличие данных
if len(z) == 0:
    print("Ошибка: нет данных для обработки. Проверьте входные данные.")
else:
    # Определение диапазона для фильтрации
    start = 100  # Начальный индекс (можно изменить)
    end = 200    # Конечный индекс (можно изменить)

    # Вырезаем подмножество данных
    z_subset = z[start:end]

    # Определение функции фильтра Калмана
    def kalman_filter(z, Q=1e-5, R=0.1):
        n = len(z)
        xhat = np.zeros(n)  # Оценка состояний
        P = np.zeros(n)     # Ковариация предсказаний
        xhat[0] = z[0]      # Начальное состояние
        P[0] = 1.0          # Начальная ошибка

        for k in range(1, n):
            # Прогнозирование
            xhat[k] = xhat[k-1]
            P[k] = P[k-1] + Q

            # Обновление
            K = P[k] / (P[k] + R)  # Коэффициент Калмана
            xhat[k] = xhat[k] + K * (z[k] - xhat[k])  # Обновленная оценка
            P[k] = (1 - K) * P[k]  # Обновление ковариации

        return xhat, P

    # Применение фильтра к очищенным данным
    filtered_data, P = kalman_filter(z_subset)

    # Рассчет стандартного отклонения
    std_dev = np.sqrt(P)  # Стандартное отклонение

    # Визуализация результатов
    plt.figure(figsize=(15, 10))
    plt.plot(z_subset, label='Изначальные данные (выборка)', alpha=0.5)
    plt.plot(filtered_data, label='Данные после фильтрации', color='red')
    plt.fill_between(np.arange(len(z_subset)), filtered_data - 3 * std_dev, filtered_data + 3 * std_dev, color='yellow', alpha=0.3, label='Интервал 3σ')

    plt.title('Применение фильтра Калмана на выбраной выборке данных')
    plt.xlabel('Индекс')
    plt.ylabel('Значение')
    plt.legend()
    plt.grid()
    plt.show()
```
